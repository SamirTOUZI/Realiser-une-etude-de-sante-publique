### Proportion de personnes en état de sous-nutrition en 2017

```python
#Selection des dataframes(données):les informations nécéssaires pour répondre se trouve dans  "population" et "sous_nutrition"
#préparation des dataframes pour pouvoir faire le calcule 
```

```python
sous_nutrition= pd.read_csv('sous_nutrition.csv', sep=',')
#plusieurs modifications à faire: selectionner dans un nouveau dataframe la population en malnutrition 2016-2018
sous_nutrition2017 =sous_nutrition.loc[sous_nutrition['Année']=='2016-2018']
#ramplacer toutes les valeurs "NAN" par 0
sous_nutrition2017 = sous_nutrition2017.fillna('0')
#ramplacer les valeurs '<0.1' par 0 
sous_nutrition2017.replace({'<0.1':'0'}, inplace = True)
#remplacer les valeurs de la colonne '2016-2018' par '2017'
sous_nutrition2017['Année']=2017
#modifier le type de la colonne valeur pour le mettre en integer 
sous_nutrition2017['Valeur'] = pd.to_numeric (sous_nutrition2017['Valeur'])
sous_nutrition2017['pop_sous_nutrition']= sous_nutrition2017['Valeur']*1000000
sous_nutrition2017.head()
```

```python
#selectionner dans un nouveau dataframe les valeurs qui correspond  à la population de l'année 2017
population2017= population.loc[population['Année']==2017]
population2017.head()
#créer une nouvelle colonne 'nb_hab' qui correspond à la colonne nbre_habitant en million que l'on convertie en millier 
population2017.insert(3,'nb_hab', population2017['nbre_habitant'].values*1000)
population2017.head()
#suppression de la colonne nbre_habitant qui est en million pour garder la colonne nb_hab en millier 
population2017.drop(columns=['nbre_habitant'], inplace=True)
population2017.head()
```

```python
# creation d'un dataframe malnutrition avec les éléments de la table population2017 et sous_nutrition2017 avec une jointure
malnutrition = pd.merge(population2017, sous_nutrition2017, on='Zone', how='inner')
malnutrition.head()
```

```python
#calcul de la somme du nombre de la population en sous nutrition et de la somme du nombre de la population 
val_sous_nutri= malnutrition['pop_sous_nutrition'].sum()
val_total= malnutrition['nb_hab'].sum()
#division du nombre total de la population en sous_nutrition par le nombre total de la population
val_total,val_sous_nutri,val_sous_nutri/val_total*100
```

### Le nombre théorique de personnes qui pourraient être nourries en 2017

```python
#Séléction de la bonne dataframe où trouver les informations nécéssaires 
dispo_alimentaire.head()
```

```python
# On regroupe part pays le nombre total de Disponibilité alimentaire 
dispo_pays_jour = dispo_alimentaire.groupby('Zone').sum()
```

```python
dispo_pays_jour.head()
```

```python
# On ne retient que les éléments de la colonne disponibilité alimentaire car se sont les seules données dont nous avons besoin
dispo_pays_jour=dispo_pays_jour['Disponibilité alimentaire (Kcal/personne/jour)']
```

```python
# creation d'un nouveau dataframe qui réuni la disponiblité alimentaire par habitant et la population total de chaques pays 
dispo_total_pays =pd.merge(dispo_pays_jour,population2017,on='Zone',how='inner')
```

```python
#on renomme la colonne car nous voulons la disponibilité alimentaire par an 
dispo_total_pays.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)':'Disponibilité alimentaire (Kcal/personne/an)'},inplace=True)
dispo_total_pays.head()
```

```python
#On multiplie la colonne par 365 jours afin d'avoir la disponibilité à l'année 
dispo_total_pays['Disponibilité alimentaire (Kcal/personne/an)']*=365
dispo_total_pays.head()
```

```python
#creation d'une nouvelle colonne qui correspond à la mutliplication de la dispo alimentaire par an avec le nombre d'habitant 
dispo_total_pays['dispo_alimentaire_Kcal_pays']= dispo_total_pays['Disponibilité alimentaire (Kcal/personne/an)']*dispo_total_pays['nb_hab']
```

```python
dispo_total_pays.head()
```

```python
#on additionne la dispo alimentaire de tous les pays pour avoir la disponibilité mondiale 
dispo_total_pays['dispo_alimentaire_Kcal_pays'].sum()
```

```python
#cette disponibilité alimentaire mondiale on la divise par le nombre de kcal necessaire par personne et par an 
dispo_total_pays['dispo_alimentaire_Kcal_pays'].sum()/(2500*365)
```

### Utilisation de la disponibilité interieure entre la part pour l'animale, la part perdue et la part pour l'homme

```python
# ON Additionne la disponibilité intérieure de chaques pays 
dispo_interieur = dispo_alimentaire.groupby('Zone').sum()
```

```python
dispo_interieur.head()
```

```python
#On additionne la disponibilité interieure de tous les pays pour avoir la disponibilité interieure mondiale 
dispo_interieur_total = dispo_interieur['Disponibilité intérieure'].sum()
```

```python
# On divise la somme des aliments pour animaux de tous les pays par la disponibilité intérieure mondiale qu'on multiplie par 100 
dispo_animaux = dispo_interieur['Aliments pour animaux'].sum()/dispo_interieur_total*100
```

```python
dispo_animaux
```

```python
#On divise la somme des pertes de tous les pays par la disponibilité intérieure mondiale qu'on multiplie par 100 
pertes = dispo_interieur['Pertes'].sum()/dispo_interieur_total*100
```

```python
pertes
```

```python
#On divise la somme de la nourriture de tous les pays par la disponibilité intérieure mondiale qu'on multiplie par 100
dispo_humain= dispo_interieur['Nourriture'].sum()/dispo_interieur_total*100
```
